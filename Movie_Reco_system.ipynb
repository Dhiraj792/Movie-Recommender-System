```python
#[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]
#['Action','Adventire','Fantasy','Science Fiction']
import ast
```

```python
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l;
```

```python
convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
def convert3(obj):
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            l.append(i['name'])
            counter+=1
        else:
            break
    return l
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
def fetch_director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
            break
    return l;
```

```python
movies['crew'].apply(fetch_director)
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
ps.stem('loved')
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
features=cv.get_feature_names_out()
```

```python
features[0:100]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
```

```python
similarity[0]
```

```python
new_df[new_df['title']=='Avatar'].index[0]
```

```python
sorted(similarity[0],reverse=True)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]#holding value of index
```

```python
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]#holding value of index
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
